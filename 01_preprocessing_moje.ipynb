```python
import pandas as pd

data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data)
df_raw
```

Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

Zmiana typu danych i wstępna eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

Podstawowe statystyki

```python
df.describe() # tylko dla kolum numerycznych
```

```python
df.describe(include=['category']).T # T za transpozycję
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# le.fit(df['bought'])
# le.transform(df['bought'])

le.fit_transform(df['bought'])
```

```python
le.classes_ # wyświetla klas które zostały zmapowane
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False) # zwraca macierz rzadką
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']]) # macierz numpy array 5x3 dla kategorii czyli tutaj rozmiarów 
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

Standaryzacja - StandardScaler

std() - pandas nieobciążony

std() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
# odchylenie standardowe jest estymatorem obciążonym wariancji
```

```python
# standaryzacja
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
# własna funkcja

def standardize(x):
  return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])

# wartości się różnią bo obciążone in nieobciążone
```

```python
# lepiej to używać
from sklearn.preprocessing import StandardScaler

encoder = StandardScaler()
encoder.fit(df[['price']])
encoder.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
